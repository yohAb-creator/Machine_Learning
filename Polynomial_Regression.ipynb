```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data preprocessing
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1: -1].values
y = df.iloc[:, -1].values
```

```python
# Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
Lin_regressor = LinearRegression()
Lin_regressor.fit(X_train, y_train)
```

```python
# Polynomial data preprocessing
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
print(X)
print(X_poly)
```

```python
# Visualizing Linear Regression
plt.scatter(X, y, color = 'red')
plt.plot(X, Lin_regressor.predict(X), "b-")
plt.title("Linear Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizating Polynomial Regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), "b-")
plt.title("Polynomial Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
# Smoother Curve
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), "b-")
plt.title("Polynomial Regression with Smoother Curves")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
# Linear Predicition
print(Lin_regressor.predict([[6.5]]))
```

```python
# Polynomial Predicition
lin_reg_2.fit(X_poly, y)
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
