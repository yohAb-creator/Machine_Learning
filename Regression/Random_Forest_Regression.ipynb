```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read and classify data
db = pd.read_csv("Position_Salaries.csv")
X = db.iloc[:, 1:-1].values
y = db.iloc[:, -1].values
```

```python
# train RFR on the whole dataset 
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X,y)
```

```python
# Predicting results
regressor.predict([[6.5]])
```

```python
# Visualizing the regression
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), 'b-')
plt.title("First Random Forest Regression")
plt.xlabel("Positions")
plt.ylabel("Salaries")
plt.show()
```

```python
# Higher Resolution Visualization
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), 'b-')
plt.title("Decision Tree Regression")
plt.xlabel("Position")
plt.ylabel("Salaries")
plt.show()
```
