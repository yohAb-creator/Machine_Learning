```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing dataset and splitting it into features and a dependent variable vector
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1: -1].values
y = df.iloc[:, -1].values
```

```python
# Training DTR on dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X,y)
```

```python
# Visualizing DTR(Result does not make sense as we are supposed to regions where features share an average value)
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), 'b-')
plt.title("Decision Tree Regression")
plt.xlabel("Position")
plt.ylabel("Salaries")
plt.show()
```

```python
# Higher Resolution Visualization
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), 'b-')
plt.title("Decision Tree Regression")
plt.xlabel("Position")
plt.ylabel("Salaries")
plt.show()
```
